```python
import imaplib
import email
import pickle

with open("email_model.pkl", "rb") as f:
    model, cv = pickle.load(f)
```

```python
user, password = "mlproject026@gmail.com","bcrb yfeq juzl evus"
# "askt djli ybqo rtmx"

# URL for IMAP connection
imap_url = 'imap.gmail.com'

# Connection with GMAIL using SSL
my_mail = imaplib.IMAP4_SSL(imap_url)

# Log in using your credentials
my_mail.login(user, password)

# Select the Inbox to fetch messages
temp = -1

while True:
    my_mail.select('Inbox')
    
    # Define Key and Value for email search
    key = 'FROM'
    _, data = my_mail.search(None, 'ALL')  # Search for emails with specific key and value
    
    id = data[0].split()[-1]
    if id == temp: 
        continue
    else:
        temp = id
        
    typ, data = my_mail.fetch(id, '(RFC822)')
    msg = data
    
    for response_part in msg:
        if type(response_part) is tuple:
            my_msg = email.message_from_bytes((response_part[1]))
            for part in my_msg.walk():
                if part.get_content_type() == 'text/plain':
                    subject = "Subject: " + my_msg['subject']
                    body = part.get_payload()
    
    # Print subject and body on separate lines
    
    l = [subject + "\n" + body]
    a = cv.transform(l)
    b = model.predict(a)
    
    if b[0] == 0:
        print("This was not a spam mail")
        print(l[0])
    else:
        print("This was a spam mail")
        print(l[0])
    
    print("%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%")
```
